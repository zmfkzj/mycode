```python
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
img = Image.open('z:/07. /10.22_// 1250 9.3m/IMG_03_1634874437_0000000099_20211022124859_586_00000018_19318211_00.jpg')
img = np.array(img)
dst = img.copy()
plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray,cmap='gray')
```

```python
canny = cv2.Canny(gray,100,100,apertureSize=3,L2gradient=True)
plt.figure(figsize=(10,10))
plt.imshow(canny)
```

```python
lines = cv2.HoughLinesP(canny, 0.8, np.pi / 180, 90, minLineLength = 2000, maxLineGap = 100)
```

```python
for i in lines:
    cv2.line(dst, (i[0][0], i[0][1]), (i[0][2], i[0][3]), (0, 0, 255), 2)

plt.figure(figsize=(10,10))
plt.imshow(dst)
```
